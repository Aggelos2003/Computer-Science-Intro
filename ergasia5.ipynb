```python
f = open('two_cities_ascii.txt', 'r') # 'r' = read
text = f.read()
print(text)
lower_text = text.lower()
print(lower_text)
splitted_text = lower_text.split()
print(splitted_text)
word_dict = {}
for word in splitted_text:
    if word in word_dict:
        word_dict[word] = word_dict[word] + 1
    else:
        word_dict[word] = 1
print('word_dict', word_dict)
sorted_word_dict = {k: v for k, v in sorted(word_dict.items(), key=lambda item: item[1], reverse=True)}
print('sorted_word_dict', sorted_word_dict)
count = 0
word_lst = []
for key in sorted_word_dict:
    word_lst.append(key)
    count += 1
    if count == 10:
        print(word_lst)
 first_2_dict = {}
for word in splitted_text:
    first_2 = word[0:2]
    if first_2 in first_2_dict:
        first_2_dict[first_2] += 1
    else:
        first_2_dict[first_2] = 1

sorted_first_2_dict = {k: v for k, v in sorted(first_2_dict.items(), key=lambda item: item[1], reverse=True)}

count = 0
first_2_lst = []
for word in sorted_first_2_dict:
    first_2_lst.append(word)
    count += 1
    if count == 3:
        print(first_2_lst)
        break
first_3_dict = {}
for word in splitted_text:
    first_3 = word[0:3]
    if first_3 in first_3_dict:
        first_3_dict[first_3] += 1
    else:
        first_3_dict[first_3] = 1

sorted_first_3_dict = {k: v for k, v in sorted(first_3_dict.items(), key=lambda item: item[1], reverse=True)}

count = 0
first_3_lst = []
for word in sorted_first_3_dict:
    first_3_lst.append(word)
    count += 1
    if count == 3:
        print(first_3_lst)
        break
```
